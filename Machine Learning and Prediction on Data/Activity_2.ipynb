```r
library(class)
library(dplyr)
```

```r
data = read.csv("german_credit_data1.csv")
```

```r
data
```

### We choose columns from Age, Credit.Amount and Job as training sets. We choose column Credit.Risks as label for test set.

### We take the sampling rate here as 0.8 below

### We take the misclassification rate for 20 values of k and we observe that k = 11 gives the lowest Misclassification rate of 0.28. Therefore we take k=11 as k that gives the highest accuracy.

```r
n.points <- 1000
sampling.rate <- 0.8
num.test.set.labels <- n.points * (1 - sampling.rate)
training <- sample(1:n.points, sampling.rate * n.points,replace=FALSE)
train <- subset(data[training, ], select = c(Age, Credit.amount,Job))
testing <- setdiff(1:n.points, training)
test <- subset(data[testing, ], select = c(Age, Credit.amount,Job))
cl <- data$Credit.Risks[training]
true.labels <- data$Credit.Risks[testing]
Misclassificationrate = c()
Kvalue = c()
for (k in 1:20) {
print(k)
predicted.labels <- knn(train, test, cl, k)
# We're using the R function knn()
num.incorrect.labels <- sum(predicted.labels != true.labels)
misclassification.rate <- num.incorrect.labels/num.test.set.labels
print(misclassification.rate)
Misclassificationrate <- c(Misclassificationrate,misclassification.rate)
Kvalue = c(Kvalue, k)
}

dt1 <- data.frame(Kvalue,Misclassificationrate)
```

```r
rs1 <- dt1 %>% filter(Misclassificationrate == min(Misclassificationrate))
rs1
```

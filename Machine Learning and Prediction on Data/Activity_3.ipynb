```r
library(dplyr)
```

```r
data = read.csv("June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv")
```

```r
a1 <- data %>% select(inc,sex,age,game1,emptype1,emplnw)
a1 <- a1 %>% filter(game1 %in% c("1","2"),inc != 99,inc != 98,age != 99,age != 98,sex %in% c("1","2"),emptype1 %in% c("1","2","3","4","5","6"),emplnw != 98,emplnw != 99)
b1 <- a1 %>% select(Income = inc,Sex = sex,Age = age,game1,emplnw)
```

## In this 5D kmeans clustering process we infer the following points - 


### Cluster 1 - Cluster size is 172 and the sum of squares is 2145.151. The sum of squares is the small and thus is a good cluster.  

### Cluster 2 - Cluster size is 202 and the sum of squares is 2815.851. The sum of squares is relatively larger and thus is not a good cluster.

### Cluster 3 - Cluster size is 145 and the sum of squares is 3969.614. This is the cluster with the smallest number of points but the sum of squares is the highest and thus is not a good cluster as the points are relatively apart.

### Cluster 4 - Cluster size is 228 and the sum of squares is 2978.053. This is the cluster with the largest number of points. The sum of squares is relatively small and thus is a good cluster.

### Cluster 5 - Cluster size is 184 and the sum of squares is 2615.940. The sum of squares is the small and thus is a good cluster.

```r
k1 <- kmeans(b1, 5, iter.max = 10, nstart = 1,algorithm = c("Hartigan-Wong", "Lloyd", "Forgy","MacQueen"))
k1
```

## In this 3D kmeans clustering process we infer the following points - 

### Cluster 1 - Cluster size is 277 and the sum of squares is 7751.227. This is the cluster with the smallest number of points. The sum of squares is the small and thus is the a good cluster.

### Cluster 2- Cluster size is 353 and the sum of squares is 14368.000. This is the cluster with the largest number of points. The sum of squares is large and thus is not a good cluster.

### Cluster 3- Cluster size is 301 and the sum of squares is 8132.053. The sum of squares is small and thus is a good cluster.

```r
k2 <- kmeans(b1, 3, iter.max = 10, nstart = 1,algorithm = c("Hartigan-Wong", "Lloyd", "Forgy","MacQueen"))
k2
```

```r
y1 = a1$emptype1
print("k=3")
table(y1,k1$cluster)
print("k=5")
y2 = a1$emptype1
table(y2,k2$cluster)
```

## Cluster plot for K = 5 cluster between Age and Income

```r
plot(b1[c("Income", "Age")], col=k1$cluster)
points(k1$centers[,c("Income", "Age")], col=1:5, pch=23, cex=3)
```

## Cluster plot for K = 3 cluster between Age and Income

```r
plot(b1[c("Income", "Age")], col=k2$cluster)
points(k2$centers[,c("Income", "Age")], col=1:3, pch=23, cex=3)
```

## Cluster plot for K = 5 cluster between Age and emplnw

```r
plot(b1[c("emplnw","Age")], col=k1$cluster)
points(k1$centers[,c("emplnw", "Age")], col=1:5, pch=23, cex=3)
```

## Cluster plot for K = 3 cluster between Age and emplnw

```r
plot(b1[c("emplnw","Age")], col=k2$cluster)
points(k2$centers[,c("emplnw", "Age")], col=1:3, pch=23, cex=3)
```
